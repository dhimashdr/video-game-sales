## Exploratory Data Analysis

**Questions need to answer**
1. How is the total game sales per genre in descending order?
2. How is the total game sales in NA, EU, JP, and other sales from 2010 to 2016?

```python
by_genre = vgsales_df.groupby(by='Genre').Global_Sales.sum().reset_index(name="Total_Sales")
by_genre
```

```python
by_year = vgsales_df.groupby(by='Year').agg({
    'NA_Sales': 'sum',
    'EU_Sales': 'sum',
    'JP_Sales': 'sum',
    'Other_Sales': 'sum'
}).reset_index()
by_year
```

```python
by_year_2010 = by_year[(by_year.Year >= "2010-01-01")]
by_year_2010_2016 = by_year_2010[(by_year_2010.Year <= "2016-01-01")]
by_year_2010_2016
```

## Data Visualization

#### 1.  How is the total game sales per genre in descending order?

```python
plt.figure(figsize=(14,5))

sns.barplot(
    data = by_genre.sort_values(by="Total_Sales", ascending=False),
    x = 'Genre',
    y = 'Total_Sales'
)

plt.title('Total Sales by Genre (1980-2020)')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.show()
```

#### 2. How is the total game sales in NA, EU, JP, and other sales from 2010 to 2016?

```python
by_year_2010_2016.plot(x="Year", y=["EU_Sales", "NA_Sales", "JP_Sales", "Other_Sales"], kind="bar", figsize=(12,4))

plt.title('Total Sales per Region (2010-2016)')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

## Conclusion

#### 1. How is the total game sales per genre in descending order?

The bar chart shows that games with action genre are became the best selling games. Meanwhile, games with strategy genre became worst-selling games.

#### 2. How is the total game sales in NA, EU, JP, and other sales from 2010 to 2016?

From the bar chart, we can see that the game sales for each regions are tend to decreasing year by year.
